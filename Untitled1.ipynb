```python
import torch
from transformers import AutoModel, AutoTokenizer, BertTokenizer

torch.set_grad_enabled(False)
```

```python
# Store the model we want to use
MODEL_NAME = "bert-large-cased"

# We need to create the model and tokenizer
model = AutoModel.from_pretrained(MODEL_NAME)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
from sklearn.manifold import TSNE 
from sklearn.decomposition import PCA
```

```python
import pandas as pd 
import numpy as np
```

```python
sent_details = pd.read_csv('eng_sentences.tsv', 
                           sep='\t', header=None, error_bad_lines=False)
```

```python
sent_details[sent_details[2].str.contains("funny")][2].values.tolist()[:3] 
```

```python
#sentences = ["This is a funny example", "This is a funny joke", "This smells funny", "Funny", "funny"]  
sentences_batch = sent_details[sent_details[2].str.contains("funny")][2].values.tolist()[:1000]
```

```python
sentences_batch = [s for s in sentences_batch if len(s) > 40 ] 
len(sentences_batch)
```

```python
def get_word_indices(tokens, word): 
  token_id = tokenizer.convert_tokens_to_ids([word])[0]   
  
  
  indices = (tokens == token_id).nonzero()  

  return indices   
```

```python
device = "cpu"
```

```python
all_outputs=[] 
step = 50 
end =10 
gpu_model = model.to(device)
for i in range(end): 
  sentences = sentences_batch[i*step:(i+1)*step]  
  print(len(sentences), "sentences")
  tokens_pt = tokenizer(sentences, return_tensors="pt", padding = True)   
  gpu_tokens  = tokens_pt.to(device)
  word_indices = None
  indices = get_word_indices(tokens_pt["input_ids"], "funny")   
  #tens[indices[:,0], indices[:,1]] 
  outputs = gpu_model(**gpu_tokens)  
  print(len(indices), "indices")
  last_hidden_state = outputs.last_hidden_state.cpu()
  last_hidden_state[indices[:,0], indices[:,1]].shape   
  
  word_outputs = last_hidden_state[indices[:,0], indices[:,1]].numpy() 
  print(len(word_outputs))

  all_outputs.append(word_outputs)  
  if len(sentences) < step:  
    #exceeded the number of sentences in batch
    break 
emb_size= all_outputs[0][0].shape[0]
all_outputs = np.concatenate(all_outputs,axis=0).reshape(-1,emb_size) 
#pickle.dump( all_outputs, open( "outputs.p", "wb" ) )
```

```python
def get_word_indices(tokens, words):  
  if type(words) == list:    
    
    token_ids = tokenizer.convert_tokens_to_ids(words)      
  else: 
    token_ids = tokenizer.convert_tokens_to_ids([words])
  
  indices = sum(tokens==token_id for token_id in token_ids).bool().nonzero() 

  return indices   
```

```python
def pipeline(sentences, words = [], indices_length=[]):  
  all_outputs=[] 
  step = 100 
  end =10 
  gpu_model = model.to(device)
  for i in range(end): 
    sentences = shorter_sentences_batch[i*step:(i+1)*step]  
    print(len(sentences))
    tokens_pt = tokenizer(sentences, return_tensors="pt", padding = True)    
    print(len(tokens_pt))
    gpu_tokens  = tokens_pt.to(device)
    word_indices = None
    indices = get_word_indices(tokens_pt["input_ids"], ["funny", "strange", "amusing"])     
    #indices = torch.cat([indices, torch.tensor([[85,3],[86,3]])]) #manual alignment for not "funny " sentences
    print(len(indices))
    #tens[indices[:,0], indices[:,1]] 
    outputs = gpu_model(**gpu_tokens)  
    
    last_hidden_state = outputs.last_hidden_state.cpu()
    last_hidden_state[indices[:,0], indices[:,1]].shape   
    
    word_outputs = last_hidden_state[indices[:,0], indices[:,1]].numpy() 

    all_outputs.append(word_outputs)  
    if len(sentences) < step:  
      #exceeded the number of sentences in batch
      break 
  emb_size = all_outputs[0][0].shape[0]
  all_outputs = np.concatenate(all_outputs,axis=0).reshape(-1,emb_size)  
  col_sentences = []
  for index in indices[:,0]: 
    col_sentences.append(shorter_sentences_batch[index.item()])
    
  total_length = step * end
  pca = PCA(n_components=2)
  pca_result = pca.fit_transform(all_outputs) 
  print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))


  pca = pd.DataFrame({"a":pca_result[:,0], "b": pca_result[:,1]})

  pca_index =pca.reset_index()
  pca_index["sentences"] = col_sentences   
  adder = 0 
  for index_length in indices_length: 
    adder += index_length.shape[0]   
  first_index_length = len(col_sentences) - adder   
  print(first_index_length) 
  print(adder)
  color_index = np.repeat(0, first_index_length) 
  for index_lengthn in indices_length: 
    color_index = np.concatenate([color_index, index_length]) 

  color_index = np.concatenate([np.repeat(0,51),np.repeat(1,33), np.repeat(2,6)])
  pca_index["color"] = color_index  

  fig = px.scatter( data_frame=pca_index, x="a", y="b", hover_data=["index", "sentences"],color="color", width=1000, height=400)
  fig.show()
  tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
  tsne_results = tsne.fit_transform(all_outputs) 
  pca_index["tsne_x"] = tsne_results[:,0] 
  pca_index["tsne_y"] = tsne_results[:,1]  
  fig = px.scatter( data_frame=pca_index, x="tsne_x", y="tsne_y", hover_data=["index", "sentences"],color="color", width=1000, height=400)
  fig.show() 
  return all_outputs

```
